Create functions for reading the corpus from TEI xml format and selecting only interesting parts.

```
import xml.etree.ElementTree as ET
import os
import glob
import gzip

def read_one(one_p):
    return ' '.join([i.text for i in one_p.findall('.//{http://www.tei-c.org/ns/1.0}f[@name="base"]/{http://www.tei-c.org/ns/1.0}string')])


def read_xml_gz(path):
  input = gzip.open(path, 'r')
  tree = ET.parse(input)
  # tree = ET.parse(path)
  root = tree.getroot()
  out = []
  for i, one_p in enumerate(root.findall('.//{http://www.tei-c.org/ns/1.0}p')):
      
      out.append(read_one(one_p))
  return out
```

Running preprocessing for each xml file. This takes around 8h.

```

for i in range(1, len(years_list)):
# for i in range(1):
  print('running years: ' + years_list[i])
  xml_files_list = [name for name in glob.glob(years_list[i]+'/**', recursive=True) if 'ann_morphosyntax.xml.gz' in name]
  print('number of files in years: ' + str(len(xml_files_list)))

  for id, file in enumerate(xml_files_list):
    if id % 20 ==0:
      print(f'current id: {id}')
    out_file_name = f'{years_list[i]}_{id}.txt'
    # print(out_file_name)
    
    temp_out = read_xml_gz(file)
    with open(output_path+out_file_name, "w") as outfile:
      outfile.write("\n".join(temp_out))
```
